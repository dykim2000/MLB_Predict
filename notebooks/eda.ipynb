```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pitching = pd.read_csv("../data/pitching_stats_2024.csv")
batting = pd.read_csv("../data/batting_stats_2024.csv")
game_logs = pd.read_csv("../data/game_logs_2024.csv")
team_batting = pd.read_csv("../data/team_batting_stats_2024.csv")
```

```python
game_logs.head(3).transpose()
```

```python
plt.figure(figsize=(12,6))
pitching_numeric = pitching.select_dtypes(include=['number'])

pitching_corr_table = pitching_numeric.corr()
#sns.heatmap(pitching_numeric.corr(), cmap='coolwarm')
pitching_corr_table['W'].sort_values(ascending=False).plot(kind="bar")
```

```python
# Features that had a high correlation with the pitcher's wins
pitching_corr_table[np.abs(pitching_corr_table['W']) > 0.5].index.tolist()
```

```python
team_batting_corr_table = team_batting.select_dtypes(include=['number']).corr()

plt.figure(figsize=(12,6))
sns.heatmap(team_batting_corr_table, cmap='coolwarm')
```
