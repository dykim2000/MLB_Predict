```python
import pandas as pd
```

```python
df_2024 = pd.read_csv("../data/final_game_logs.csv")
df_2025 = pd.read_csv("../data/final_game_logs_2025.csv")
```

```python
print("2024 columns:", len(df_2024.columns))
print("2025 columns:", len(df_2025.columns))

missing_in_2025 = set(df_2024.columns) - set(df_2025.columns)
extra_in_2025 = set(df_2025.columns) - set(df_2024.columns)

print("Missing in 2025:", missing_in_2025)
print("Extra in 2025:", extra_in_2025)
```

```python
cols_to_drop = ["Unnamed: 0", "temp_y", "condition_y", "wind_y"]

df_2024 = df_2024.drop(columns=cols_to_drop, errors="ignore")
df_2025 = df_2025.drop(columns=cols_to_drop, errors="ignore")
```

```python
df_2024.to_csv("../data/final_game_logs.csv", index=False)
df_2025.to_csv("../data/final_game_logs_2025.csv", index=False)
```
