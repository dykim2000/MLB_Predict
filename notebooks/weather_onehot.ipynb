## One-Hot encoding the Weather Condition Column

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/final_game_logs_2024.csv")
```

```python
df["condition_code"] = df["condition_code"].astype("category")
condition_dummies = pd.get_dummies(df["condition_code"], prefix="condition")
```

```python
df = pd.concat([df, condition_dummies], axis=1)
df = df.drop(columns=["condition_code"])
```

```python
df = df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"])
```
